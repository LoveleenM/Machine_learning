```python
import cv2
```

```python
im=cv2.imread("./elephant.jpg")
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(im)
```

```python
print(im.shape)
```

```python
im2=im.reshape((im.shape[0]*im.shape[1],im.shape[2]))
```

```python
im2.shape
```

```python
im3=im.reshape((-1,im.shape[2]))
```

```python
im3.shape
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_color=5
```

```python
km=KMeans(n_clusters=dominant_color)
```

```python
km.fit(im2)
```

```python
center=km.cluster_centers_
```

```python
import numpy as np
```

```python
centers=np.array(center,dtype='uint8')
```

```python
print(centers)
```

```python
i=1
plt.figure(0,figsize=(4,4))#height and width
cente=[]
for all1 in centers:
    plt.subplot(1,5,i)
    plt.axis("off")
    i+=1
    cente.append(all1)
    
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=all1
    plt.imshow(a)
plt.show()
    
```

```python
new_img=np.zeros((im2.shape[0],im2.shape[1]),dtype='uint8')
```

```python
new_img.shape
```

```python
cente
```

```python
km.labels_
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix]=cente[km.labels_[ix]]
new_img=new_img.reshape(im.shape[0],im.shape[1],im.shape[2])

plt.imshow(new_img)
plt.show()
```
