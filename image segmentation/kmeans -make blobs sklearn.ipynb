```python
k=5
color=["green","orange","red","yellow","brown"]
clusters={}
for kx in range(k):
    center=10*(np.random.random((x.shape[1],)))
    points=[]
    cluster={
        "points":points,
        "center":center,
        "color":color[kx]
    }
    clusters[kx]=cluster
    
    
```

```python
def distance(v1,v2):
    return(np.sqrt(np.sum(v1-v2)**2))
```

```python
def assignpoints(clusters):
    for i in range(x.shape[0]):
        c=x[i]
        dist=[]
        for kx in range(k):
            d= distance(c,clusters[kx]['center'])
            dist.append(d)
        index=np.argmin(dist)#returns index of min distance
        clusters[index]['points'].append(c)
    
```

```python
clusters
```

```python
def updateClusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        print(pts)
        if pts.shape[0]>1:
            
            clusters[kx]['center']=pts.mean(axis=0)#vertically
            clusters[kx]['points']=[]
        print(clusters[kx]['center'])
```

```python
updateClusters(clusters)
```

```python
import matplotlib.pyplot as plt
```

```python
def plotclusters(clusters):
    for kx in range(k):
        #print(clusters[kx]['points'])
        pts=np.array(clusters[kx]['points'])
        print(pts.shape)
        try:
            print("hi")
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk=clusters[kx]['center']
        print(uk)
        plt.scatter(uk[0],uk[1],color="black",marker="*")
```

```python
plotclusters(clusters)
```

```python
assignpoints(clusters)
```

```python
plotclusters(clusters)
```

```python
updateClusters(clusters)
```
