```python
query_x=np.array([2,3])
plt.scatter(query_x[0],query_x[1],color='red')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(X,Y,querypoint,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=dist(querypoint,X[i])
        vals.append((d,Y[i]))
        
        
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    new_val=np.unique(vals[:,1],return_counts=False)
    print(new_val)
    return vals
```

```python
knn(X,Y,query_x)
```
