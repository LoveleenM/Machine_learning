```python
def generateTable(data,k=4):
    T = {}
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        #print("T.get(X)",T.get(X))#returns value for a certain key
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
            #print(T)
        else:
            #print("T[X].get(Y)",T[X].get(Y))
            if T[X].get(Y) is None:
                T[X][Y] =1
            else:
                T[X][Y] += 1
            #print(T)
    return T
            
        
```

```python
def convertFreqIntoProb(T):
    for kx in T.keys():
        #print(kx)
        s=float(sum(T[kx].values()))
        #print(T[kx])
        for k in T[kx].keys():
            T[kx][k]=T[kx][k]/s
    return T
```
