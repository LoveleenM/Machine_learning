```python
def generateTable(data,k=4):
    T={}
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        
        if T.get(X) is None:
            T[X]={}
            T[X][Y]=1
        else:
            if T.get(Y) is None:
                T[X][Y]=1
            else:
                T[X][Y]+=1
    return T
```

```python
def convertFreqIntoProb(T):
    for kx in T.keys():
        s=float(sum(T[kx].values()))
        for k in T[kx].keys():
            T[kx][k]=T[kx][k]/s
        
    return T
```

```python
def train(data,k=4):
    T=generateTable(data,k)
    T=convertFreqIntoProb(T)
    return T
```

```python
def sample_next(ctx,T):
    ctx=ctx[-4:]
    if T.get(ctx) is None:
        return " "
    possible_char=list(T[ctx].keys())
    possible_vals=list(T[ctx].values())
    #print(possible_vals)
    return np.random.choice(possible_char,p=possible_vals)
```

```python
def generate_Text(starting_sent,maxlen=1000):
    np.random.seed(1)#same result everytime
    sentence=starting_sent
    ctx=starting_sent[-4:]
    for ix in range(maxlen):
        next_prediction=sample_next(ctx,T)
        sentence+=next_prediction
        ctx=sentence[-4:]
    return sentence
```
