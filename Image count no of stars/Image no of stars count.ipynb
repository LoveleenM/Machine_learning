```python
from skimage.io import imread
```

```python
example_file=glob.glob(r"./sky.jpeg")[0]
print(example_file)
```

```python
im=imread(example_file,as_gray=True)
plt.imshow(im)
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=5, num_sigma=10, threshold=.1)
print(blobs_log)
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
print(blobs_log)
numrows = len(blobs_log)
print("Number of stars counted : " ,numrows)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(im)
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+1, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)
```
